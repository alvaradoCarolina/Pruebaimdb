```python
import pandas as pd

ruta_csv = 'limpio_imdb_movie_data_2023.csv'

df = pd.read_csv(ruta_csv)

df.head()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Rating'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribución de Ratings')
plt.xlabel('Rating')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(data=df, x='Meta Score', bins=30, kde=True)
plt.title('Distribución de Metas')
plt.xlabel('Meta Numero')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(14, 8))
sns.countplot(x='Year', data=df, palette='viridis')
plt.xticks(rotation=90)
plt.title('Número de Películas por Año')
plt.xlabel('Año')
plt.ylabel('Número de Películas')
plt.show()
```

```python
plt.scatter(df['Votes'], df['Rating'], alpha=0.5, color='coral')
plt.title('Relación entre Votos y Ratings')
plt.xlabel('Votos')
plt.ylabel('Rating')
plt.show()
```

```python
plt.scatter(df['Rating'], df['Meta Score'], alpha=0.5, color='green')
plt.xlabel('Rating')
plt.ylabel('Meta Score')
plt.title('Relación entre Rating y Meta Score')
plt.show()
```

```python
import seaborn as sns

sns.boxplot(x=df['Duration'])
plt.xlabel('Duración')
plt.title('Distribución de Duración de Películas')
plt.show()
```

```python
votes_by_year = df.groupby('Year')['Votes'].mean()
votes_by_year.plot(kind='line', marker='o', color='orange')
plt.xlabel('Año')
plt.ylabel('Promedio de Votos')
plt.title('Tendencia de Votos a lo largo de los Años')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['Rating'], df['Meta Score'], df['Votes'], c='blue', marker='o')
ax.set_xlabel('Rating')
ax.set_ylabel('Meta Score')
ax.set_zlabel('Votos')
ax.set_title('Rating vs. Meta Score vs. Votos')
plt.show()
```

```python
plt.scatter(df['Year'], df['Rating'], s=df['Votes']/1000, alpha=0.5, c='red', edgecolors='gray')
plt.xlabel('Año')
plt.ylabel('Rating')
plt.title('Año vs. Rating vs. Votos')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df_subset = df.head(45)

plt.figure(figsize=(18, 10))

sns.violinplot(x='Genre', y='Duration', data=df_subset, palette='viridis')

plt.xlabel('Género')
plt.ylabel('Duración')
plt.title('Distribución de Duración por Género')
plt.xticks(rotation=45)
plt.show()
```
